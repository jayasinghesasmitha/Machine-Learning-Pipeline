```python
import pandas as pd
import numpy as np
```

```python
d1 = {'Social_media_followers':[1000000, np.nan, 2000000, 1310000, 1700000, np.nan, 4100000, 1600000, 2200000, 1000000],
    'Sold_out':[1,0,0,1,0,0,0,1,0,1]}
```

```python
df1 = pd.DataFrame(data=d1)
```

```python
df1
```

```python
X1 = df1[['Social_media_followers']]
```

```python
Y1 = df1[['Sold_out']]
```

```python
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1, Y1, test_size=0.3, random_state=19)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy = 'mean')
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe1 = make_pipeline(imputer,lr)
```

```python
pipe1.fit(X1_train,Y1_train)
```

```python
pipe1.score(X1_train , Y1_train)
```

```python
pipe1.score(X1_test , Y1_test)
```

```python
d2 = {'Genre':['Rock', 'Metal', 'Bluegrass', 'Rock', np.nan, 'Rock', 'Rock', np.nan, 'Bluegrass', 'Rock'],
    'Social_media_followers':[1000000, np.nan, 2000000, 1310000, 1700000, np.nan, 4100000, 1600000, 2200000, 1000000],
    'Sold_out':[1,0,0,1,0,0,0,1,0,1]}
```

```python
df2 = pd.DataFrame(data=d2)
```

```python
df2
```

```python
X2 = df2.iloc[: , 0:2]
```

```python
Y2 = df2.iloc[: , 2]
```

```python
X2_train, X2_test, Y2_train, Y2_test = train_test_split(X2, Y2, test_size=0.3, random_state=170)
```

```python
num_cols = ['Social_media_followers']
```

```python
cat_cols = ['Genre']
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
num_pipeline = Pipeline(steps = [('impute',SimpleImputer(strategy = 'mean')) , ('scale' , StandardScaler())])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
cat_pipeline = Pipeline(steps =[('impute',SimpleImputer(strategy = 'most_frequent')), ('one-hot-encoder', OneHotEncoder(handle_unknown = 'ignore' , sparse_output = False))])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
col_transformer = ColumnTransformer(transformers = [('num_pipeline',num_pipeline,num_cols),('cat_pipeline',cat_pipeline,cat_cols)] , remainder = 'drop',n_jobs = -1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
pipefinal = make_pipeline(col_transformer , dtc)
```

```python
pipefinal.fit(X2_train , Y2_train)
```

```python
pipefinal.score(X2_test, Y2_test)
```

```python
#how to save the pipeline
```
